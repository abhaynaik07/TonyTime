```python
num_feature = [f for f in df.columns if df[f].dtype != 'O']
cat_feature = [f for f in df.columns if df[f].dtype == 'O']

print("We have {} numerical features: {}".format(len(num_feature),num_feature))

print("\nWe have {} categorical features: {}".format(len(cat_feature),cat_feature))
```

```python
df.head(2)
```

```python
df['total'] = df['math score']+df['reading score']+df['writing score']
df['average'] = df['total']/3
df.head()
```

```python
read_full = df[df['reading score'] == 100]['average'].count()
write_full =df[df['writing score'] == 100]['average'].count()
math_full =df[df['math score'] == 100]['average'].count()

print(f'no. of student with full marks in math :{math_full}')
print(f'no. of student with full marks in write :{write_full}')
print(f'no. of student with full marks in read :{read_full}')
```

```python
read_less20 = df[df['reading score'] <= 20]['average'].count()
write_less20 =df[df['writing score'] <= 20]['average'].count()
math_less20 =df[df['math score'] <= 20]['average'].count()

print(f'no. of student with full marks in math :{math_less20}')
print(f'no. of student with full marks in w :{write_less20}')
print(f'no. of student with full marks in read :{read_less20}')
```

```python
#Histogram
fig ,axs = plt.subplots(1,2, figsize=(15,7))
plt.subplot(121)
sns.histplot(data=df , x ='average',bins =30,kde=True,color='g')
plt.subplot(122)
sns.histplot(data=df , x ='average',kde=True,hue='gender')
plt.show()
```
